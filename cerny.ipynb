```python
def mapappender(s,current):
    global maps
    if len(current)<len(s):
        for i in s:
            mapappender(s,current+[i])
    else:
        maps.append(current)
    

def allmaps(s): #for list s, returns all maps s->s, with a map represented as a list of elements from s
    global maps
    maps=[]
    mapappender(s,[])
    return maps

    
```

```python
class DFA: #implementing a discrete finite automaton
    def __init__(self, transdict): #transdict should be a dictionary with keys being states of the DFA and the value at a state being another dictionary with keys being elements of the alphabet (in most cases 0 and 1) and values being the targets of edges with these labels from the given state
        self.states = set(transdict.keys())
        self.size = len(self.states)
        self.alphabet = set(list(transdict.values())[0].keys())
        self.transitions = transdict


def allcyclicDFAs(n): #outputs list of all DFAs with state set {0,1,...,n-1}, alphabet {0,1}, with the 1 transition being +1 mod n
    nlist=[]
    for i in range (n):
        nlist.append(i)
    alm=allmaps(nlist)
    
    DFAlist=[]
    for m in alm:
        trandict={}
        for i in range(n):
            trandict[i]={0:m[i],1:((i+1)%n)}
        DFAlist.append(DFA(trandict))
    
    return DFAlist


def coprime(cyclicautomaton): #checks if cyclic automaton with convention as above merges some i,j with gcd(i-j,n)=1
    for i in cyclicautomaton.states:
        for j in cyclicautomaton.states:
            if (cyclicautomaton.transitions[i][0]==cyclicautomaton.transitions[j][0]):
                if math.gcd(i-j,cyclicautomaton.size)==1:
                    return True
    
    return False
    
    

    
```

```python
def pairgraph(automaton): #outputs the unlabeled directed graph (represented by adjacency dictionary) with vtx set being pairs of states and edges being pair transitions
    g = {}
    for i in automaton.states:
        for j in automaton.states:
            g[frozenset({i,j})]={frozenset({automaton.transitions[i][0],automaton.transitions[j][0]}),frozenset({automaton.transitions[i][1],automaton.transitions[j][1]})}
    return g
```

```python
def reverse(graph): #given graph written as dictionary, outputs graph with all edges reversed (again as dictionary)
    revg={}
    for vtx in graph:
        revg[vtx]=set()
    
    for vtx in graph:
        for outvtx in graph[vtx]:
            current=revg[outvtx]
            revg[outvtx]=current.union({vtx})
    return revg

newg = reverse(g)
newg
```

```python
def dfs(visited, graph, vtx): #depth-first search, graph given as dictionary, visited as set
    if vtx not in visited:
        visited.add(vtx)
        for neighbor in graph[vtx]:
            dfs(visited, graph, neighbor)

#visited=set()
#dfs(visited,newg,frozenset({0}))
#print(visited)
#print(len(visited))
#print(len(newg))
```

```python
def synchronizing(automaton): #checks if cyclic automaton that contains state 0 is synchronizing; does not work for general automata (to make it work in general, could do dfs from all vtxs instead, but for cyclic it suffices to do it from {0} (and this is a bit faster))
    g = pairgraph(automaton)
    newg = reverse(g)
    visited = set()
    dfs(visited,newg,frozenset({0}))
    return len(newg)==len(visited)

#synchronizing(automaton)
```
